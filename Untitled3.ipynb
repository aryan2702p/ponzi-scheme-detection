```python
import re

for index, row in data.iterrows():
    opcode_string = row['opcode']
    opcodes = re.findall(r'\b[A-Z0-9]+(?=\s|$)', opcode_string)
    sorted_opcodes = sorted(set(opcodes))
    opcode_modified = ' '.join(sorted_opcodes)
    data.at[index, 'opcode'] = opcode_modified
```

```python
import re
import pickle
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

def preprocess_opcodes(data, vocab_file):
    preprocessed_opcodes = []

    # Iterate over each row in the DataFrame
    for index, row in data.iterrows():
        # Extract opcode strings from the current row
        opcodes_row = re.findall(r'\b[A-Z0-9]+(?=\s|$)', row['opcode'])
        # Remove duplicates, sort opcodes, and concatenate them into a single string
        opcodes_sorted_unique = ' '.join(sorted(set(opcodes_row)))
        data.at[index, 'opcode'] = opcodes_sorted_unique    
        preprocessed_opcodes.append(opcodes_sorted_unique)

    # Remove duplicates and sort opcodes from all rows to create vocabulary
    opcodes_all_rows = ' '.join(preprocessed_opcodes)
    opcodes_unique_sorted = sorted(set(re.findall(r'\b[A-Z0-9]+(?=\s|$)', opcodes_all_rows)))
    
    # Save vocabulary to file
    with open(vocab_file, 'wb') as f:
        pickle.dump(opcodes_unique_sorted, f)
    
    # Create a tokenizer
    tokenizer = Tokenizer()
    
    # Fit tokenizer on the vocabulary
    tokenizer.fit_on_texts(opcodes_unique_sorted)
    
    # Convert opcode text into integer sequences for each row separately
    opcode_sequences = tokenizer.texts_to_sequences(preprocessed_opcodes)
    
    # Pad sequences to ensure they have the same length
    max_sequence_length = max(len(seq) for seq in opcode_sequences)
    padded_sequences = pad_sequences(opcode_sequences, maxlen=max_sequence_length, padding='post')
    
    return padded_sequences, tokenizer, max_sequence_length

# Example usage:
vocab_file = 'opcode_vocab.pkl'
preprocessed_sequences, tokenizer,max_sequence_length = preprocess_opcodes(data, vocab_file)
print("Preprocessed sequences:", preprocessed_sequences)
print("Vocabulary size:", len(tokenizer.word_index))
print("Max length")
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data['opcode'])
X_tokenized = tokenizer.texts_to_sequences(data['opcode'])

max_sequence_length = max(len(seq) for seq in X_tokenized)
X_padded = pad_sequences(X_tokenized, maxlen=max_sequence_length)
```

```python
print(X_padded)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X_padded, data['label'], test_size=0.2, random_state=42)
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_padded, data['label'], test_size=0.2, random_state=42, stratify=data['label'])


X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42, stratify=y_train)

```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, LSTM, Dense, Embedding, Dropout, Flatten
from sklearn.model_selection import train_test_split
import numpy as np

def cnn_lstm_Model():
    model = Sequential([
        Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64, input_length=max_sequence_length),
        Conv1D(64, 5, activation='relu'),
        MaxPooling1D(pool_size=4),
        LSTM(128),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def mlp_lstm_Model():
    model = Sequential([
        Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64, input_length=max_sequence_length),
        LSTM(128, return_sequences=True),
        Flatten(),
        Dense(64, activation='relu'),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def cnn_lstm_dropout_Model():
    model = Sequential([
        Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64, input_length=max_sequence_length),
        Conv1D(64, 5, activation='relu'),
        MaxPooling1D(pool_size=4),
        LSTM(128),
        Dropout(0.5),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def cnn_lstm_flatten_Model():
    model = Sequential([
        Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64, input_length=max_sequence_length),
        Conv1D(64, 5, activation='relu'),
        MaxPooling1D(pool_size=4),
        LSTM(128),
        Flatten(),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

def test_model():
    model = Sequential([
        Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=64, input_length=max_sequence_length),
        LSTM(128),
        Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def plot_confusion_matrix(model, model_name, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_classes = (y_pred > 0.5).astype("int32")

    conf_matrix = confusion_matrix(y_test, y_pred_classes)

    accuracy = accuracy_score(y_test, y_pred_classes)
    precision = precision_score(y_test, y_pred_classes)
    recall = recall_score(y_test, y_pred_classes)
    f1 = f1_score(y_test, y_pred_classes)

    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}\nAccuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, F1-score: {f1:.2f}")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

```

```python
cnn_lstm_model = cnn_lstm_Model()
cnn_lstm_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
mlp_lstm_model = mlp_lstm_Model()
mlp_lstm_model.fit(X_train, y_train, epochs=15, batch_size=64, validation_data=(X_val, y_val))
```

```python
cnn_lstm_dropout_model = cnn_lstm_dropout_Model()
cnn_lstm_dropout_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
cnn_lstm_flatten_model = cnn_lstm_flatten_Model()
cnn_lstm_flatten_model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
plot_confusion_matrix(cnn_lstm_model, "CNN-LSTM", X_test, y_test)
plot_confusion_matrix(mlp_lstm_model, "MLP_LSTM", X_test, y_test)
plot_confusion_matrix(cnn_lstm_dropout_model, "CNN_LSTM_DROP", X_test, y_test)
plot_confusion_matrix(cnn_lstm_flatten_model, "CNN_LSTM_FLAT", X_test, y_test)
```
