```python
import pandas as pd
data = pd.read_csv('Ponzi_contracts.csv')
from keras.models import load_model
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
import re

for index, row in data.iterrows():
    opcode_string = row['opcode']
    opcodes = re.findall(r'\b[A-Z0-9]+(?=\s|$)', opcode_string)
    sorted_opcodes = sorted(set(opcodes))
    opcode_modified = ' '.join(sorted_opcodes)
    data.at[index, 'opcode'] = opcode_modified
```

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(data['opcode'])
X_tokenized = tokenizer.texts_to_sequences(data['opcode'])

max_sequence_length = max(len(seq) for seq in X_tokenized)
X_padded = pad_sequences(X_tokenized, maxlen=max_sequence_length)
```

```python
# X_train, X_test, y_train, y_test = train_test_split(X_padded, data['label'], test_size=0.2, random_state=42)
# X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)

from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(X_padded, data['label'], test_size=0.2, random_state=42, stratify=data['label'])


X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42, stratify=y_train)

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score

def plot_confusion_matrix(model, model_name, X_test, y_test):
    y_pred = model.predict(X_test)
    y_pred_classes = (y_pred > 0.5).astype("int32")

    conf_matrix = confusion_matrix(y_test, y_pred_classes)

    accuracy = accuracy_score(y_test, y_pred_classes)
    precision = precision_score(y_test, y_pred_classes)
    recall = recall_score(y_test, y_pred_classes)
    f1 = f1_score(y_test, y_pred_classes)

    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues")
    plt.title(f"Confusion Matrix - {model_name}\nAccuracy: {accuracy:.2f}, Precision: {precision:.2f}, Recall: {recall:.2f}, F1-score: {f1:.2f}")
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.show()

```

```python
plot_confusion_matrix(model, "CNN-LSTM", X_test, y_test)
```

```python
from keras.layers import Bidirectional, LSTM, Dense, Embedding, Dropout
from keras.models import Sequential

model = Sequential([
    Embedding(input_dim=len(tokenizer.word_index) + 1, output_dim=100, input_length=max_sequence_length),
    Bidirectional(LSTM(128, return_sequences=True)),
    Dropout(0.5),
    Bidirectional(LSTM(64)),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10, batch_size=64, validation_data=(X_val, y_val))
```

```python
plot_confusion_matrix(model, "CNN-LSTM", X_test, y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Assuming you have your data prepared as X_train, y_train, X_test, y_test

# Decision Tree
dt_model = DecisionTreeClassifier()
dt_model.fit(X_train, y_train)
dt_predictions = dt_model.predict(X_test)
dt_accuracy = accuracy_score(y_test, dt_predictions)
print("Decision Tree Accuracy:", dt_accuracy)

# Random Forest
rf_model = RandomForestClassifier()
rf_model.fit(X_train, y_train)
rf_predictions = rf_model.predict(X_test)
rf_accuracy = accuracy_score(y_test, rf_predictions)
print("Random Forest Accuracy:", rf_accuracy)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define a function to plot confusion matrix
def plot_confusion_matrix(conf_matrix, title):
    sns.heatmap(conf_matrix, annot=True, cmap='Blues', fmt='d', 
                xticklabels=['Predicted Negative', 'Predicted Positive'], 
                yticklabels=['Actual Negative', 'Actual Positive'])
    plt.title(title)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

# Plot confusion matrix for Naive Bayes
plot_confusion_matrix(nb_conf_matrix, 'Naive Bayes Confusion Matrix')

# Plot confusion matrix for SVM
#plot_confusion_matrix(svm_conf_matrix, 'SVM Confusion Matrix')

# Plot confusion matrix for Gradient Boosting Machines
plot_confusion_matrix(gbm_conf_matrix, 'GBM Confusion Matrix')
```
